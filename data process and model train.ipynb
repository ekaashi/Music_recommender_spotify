This is a regular expression (regex) pattern.

🔹 \s

Means any whitespace character, including:

space " "

newline \n

carriage return \r

tab \t

🔹 +

Means one or more

So:

\s+

means:

“One or more whitespace characters”

🔎 What does this line do?
.str.replace(r'\s+', ' ', regex=True)

It replaces:

multiple spaces, \n, \r, tabs

with

a single space
🔥 Example

Before:

Hello   world

This   is   text

After:

Hello world This is text

Clean and single-spaced.

🔎 Why r before the string?
r'\s+'

r means raw string.

Without r, Python may misinterpret \s as an escape character.

So always use:

r'\s+'

when writing regex.

🔎 Why regex=True?

Because we are using a regular expression pattern.

If regex=False, Pandas treats \s+ as normal text, not a pattern.

```python
df.text=df.text.replace(r'\s+',' ',regex=True)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
# it is from list comprehension
'''def token(txt):
    token = nltk.word_tokenize(txt)
    a =[stemmer.stem(w) for w in token]
    return " ".join(a)'''
```

```python
def token(txt):
    token=nltk.word_tokenize(txt)
    a=[]
    for w in token:
        a.append(stemmer.stem(w))
    return" ".join(a)
```

```python
token("u are beautifull,beauti")
```

```python
df['text'].apply(lambda x: token(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid = TfidfVectorizer(analyzer ='word',stop_words='english')
```

```python
X =tfid.fit_transform(df['text'])
```

```python
similar=cosine_similarity(X)
```

```python
def recommend(songs):
    idx= df[df['song']==songs].index[0]
    distance=sorted(list(enumerate(similar[idx])),reverse=True,key=lambda x:x[1])
    song=[]
    for s_id in distance[1:21]:
        song.append(df.iloc[s_id[0]].song)
    return song    
```

```python
recommend('Bang-A-Boomerang')
```
